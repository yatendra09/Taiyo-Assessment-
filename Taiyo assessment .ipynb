## E-procurement Government of India:

```python
url="https://etenders.gov.in/eprocure/app?page=FrontEndTendersByOrganisation&service=page"

page=requests.get(url)
soup=BeautifulSoup(page.text,"lxml")
```

```python
soup
```

```python
table1=soup.find('table',{"id":"table"})
table1
```

```python
#table
```

```python
df = pd.DataFrame(columns = ['S.No',"Organisation_Name"])
df
```

```python
for i in table1.find_all('tr',{'id':"informal"}):
    row_data = i.find_all('td',align="left")
    row = [tr.text for tr in row_data]
    df.loc[len(df)] = row
    
    
```

```python
row
```

```python
for i in range(0,60):
    for j in table1.find_all('tr',{'id':'informal_'+str(i)}):
        row_data=j.find_all('td',align='left')
        row=[tr.text for tr in row_data]
        df.loc[len(df)] = row
```

```python
len(df)
df
```

```python
tender_count=[]

for j in table1.find_all('tr',{'id':"informal"}):
    #row_data = i.find_all('td',align='right')
        tender=j.find('a').contents[0]
        tender=tender.replace("\t","")
        tender=tender.replace("\n","")
        tender=tender.replace("\r","")
        tender=tender.strip()
        
        tender_count.append(tender)
```

```python

for i in range(0,60):

    for j in table1.find_all('tr',{'id':"informal_"+str(i)}):
    #row_data = i.find_all('td',align='right')
        tender=j.find('a').contents[0]
        tender=tender.replace("\t","")
        tender=tender.replace("\n","")
        tender=tender.replace("\r","")
        tender=tender.strip()
        
        tender_count.append(tender)
```

```python
len(tender_count)
```

```python
df['Tender_count']=tender_count
df.head()
```

```python
df.to_csv("Organisation_tender_count.csv")
```
